```python
predictor_var = ['age','bmi',
 'beard',
 'mallampati',
 'neck_radiation',
 'ulbt',
 'teeth','neck_circumference','macroglossia','history_of_snoring',
 'stiff_lungs']
outcome_var='target'
model=LogisticRegression()
classification_model(model,df,traindf,testdf,predictor_var,outcome_var)
```
